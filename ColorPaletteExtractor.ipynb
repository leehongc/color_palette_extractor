```python
# Importing the necessary libraries

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
image = Image.open('images/mario.png')
```

```python
image
```

```python
reduce_factor = 10
# this is used to reduce the size of the image
image_size = image.size
w, h = image_size

image_small = image.resize((int(w/reduce_factor), int(h/reduce_factor)))
# reducing the resolution of the original image

image_small
```

```python
pixelated = image_small.resize(image.size,Image.NEAREST)
# this is to resize the small image to its original size

pixelated
```

```python
rgb_array = np.array(image_small.convert('RGB'))
# convert the small image to an RGB array

pixels = rgb_array.reshape(-1, 3)
# reshape the RGB array to a 2d array
```

```python
len(pixels)
```

```python
num_samples = 5
# sample of colors to sample

random_indices = np.random.choice(pixels.shape[0], num_samples, replace=False)
random_pixels = pixels[random_indices]
# randomly sample colors with the specified number of samples

random_pixels
```

```python
# Reshape the RGB colors array to a 2D grid
grid = random_pixels.reshape(1, -1, 3)

# Create a figure and plot the colors
fig, ax = plt.subplots(figsize=(8, 2))
ax.imshow(grid, aspect='auto')

# Customize the plot
ax.axis('off')

# Show the plot
plt.show()
```

```python
def display_random_sample(color_array):
    num_samples = 5
    # sample of colors to sample

    random_indices = np.random.choice(color_array.shape[0], num_samples, replace=False)
    random_color_array = color_array[random_indices]
    # randomly sample colors with the specified number of samples
    # Reshape the RGB colors array to a 2D grid
    grid = random_color_array.reshape(1, -1, 3)

    # Create a figure and plot the colors
    fig, ax = plt.subplots(figsize=(8, 2))
    ax.imshow(grid, aspect='auto')

    # Customize the plot
    ax.axis('off')

    # Show the plot
    plt.show()
```

```python
# testing to see how shuffling the original image multiple times would affect the color sample

rgb_array = np.array(image.convert('RGB'))
# convert the small image to an RGB array

array = rgb_array.reshape(-1, 3)
# reshape the RGB array to a 2d array

for i in range(6):
    random.shuffle(array)

display_random_sample(array)
```

```python
# testing to see how shuffling the pixelated image multiple times would affect the color sample

for i in range(6):
    random.shuffle(pixels)

display_random_sample(pixels)
```
